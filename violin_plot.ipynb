```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the datasets
data_2018_2019 = pd.read_csv('Database/final/crime_data_modified_2018_removed_final.csv')
data_2020_2022 = pd.read_csv('Database/final/crime_data_modified_final.csv')

# Combine the datasets
combined_data = pd.concat([data_2018_2019, data_2020_2022], ignore_index=True)

# Convert the 'vict_age' from string to numeric, if necessary
combined_data['vict_age'] = pd.to_numeric(combined_data['vict_age'], errors='coerce')

# Filter the dataset to include only the necessary years and valid ages
filtered_data = combined_data[combined_data['year_rept'].isin([2018, 2019, 2020, 2021, 2022])]
filtered_data = filtered_data.dropna(subset=['vict_age'])

# Create the violin plot with enhanced visual settings
plt.figure(figsize=(10, 6))
sns.violinplot(x='year_rept', y='vict_age', data=filtered_data)
plt.title("DISTRIBUTION OF VICTIM'S AGE FROM 2018 to 2022", fontsize=24, color='Black', fontname='Arial')
plt.xlabel('Year')
plt.ylabel('Victim\'s Age')
plt.grid(True, linestyle='--')
plt.gca().set_facecolor('#e6e0e1')  # Light pink background for the plot area
plt.gcf().set_facecolor('#fdc1c5')  # Light yellow background for the entire figure
plt.show()
```
